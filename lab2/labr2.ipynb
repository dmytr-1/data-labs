```python
import os
import datetime
import pandas as pd
import urllib.request
```

1. Для кожної із адміністративних одиниць України завантажити тестові
структуровані файли, що містять значення VHI-індексу. Ця процедура
має бути автоматизована, параметром процедури має бути індекс
(номер) області. При зберіганні файлу до його імені потрібно додати
дату та час завантаження;

```python
def vhi_data(province_id, start_year=2000, end_year=2024):

    storage_directory = "lab2"
    if not os.path.exists(storage_directory):
        os.makedirs(storage_directory)
    

    existing_files = []
    for file in os.listdir(storage_directory):
        if file.startswith(f"VHI_ID_{province_id}_") and file.endswith(".csv"):
            existing_files.append(file)
    
    if existing_files:
        print(f"Файл №{province_id} вже існує: {existing_files[0]}\n")
        return
    

    download_url = (
        f"https://www.star.nesdis.noaa.gov/smcd/emb/vci/VH/get_TS_admin.php?"
        f"country=UKR&provinceID={province_id}&year1={start_year}&year2={end_year}&type=Mean"
    )
    vhi_data = urllib.request.urlopen(download_url)
    

    current_date = datetime.datetime.now().strftime("%d-%m-%Y")
    current_time = datetime.datetime.now().strftime("%H-%M-%S")
    filename = f"VHI_ID_{province_id}_{current_date}_{current_time}.csv"
    

    file_path = os.path.join(storage_directory, filename)
    with open(file_path, 'wb') as output_file:
        output_file.write(vhi_data.read())
    

    print(f"\nфайл {filename} успішно завантажилось!")


for province_id in range(1, 28):
    print(f"№{province_id}...")
    vhi_data(province_id, 1982, 2024)
```

2. Зчитати завантажені текстові файли у фрейм (детальніше про роботу
із фреймами буде розказано у подальших лабораторних роботах).
Імена стовбців фрейму мають бути змістовними та легкими для
сприйняття (не повинно бути спеціалізованих символів, пробілів
тощо). Ця задача має бути реалізована у вигляді окремої процедури,
яка на вхід приймає шлях до директорії, в якій зберігаються файли;

```python
def read_files(directory):
    files = os.listdir(directory)
    headers = ['Year', 'Week', 'SMN', 'SMT', 'VCI', 'TCI', 'VHI', 'empty']
    data_frame = pd.DataFrame()
    
    for i in range(len(files)):
        file_path = os.path.join(directory, files[i])

        province_id = int(files[i].split('_')[2])
        
        if province_id in [12, 20]:
            continue
        
        df = pd.read_csv(file_path, header=1, names=headers)
        
        df = df.drop(df.loc[df['VHI'] == -1].index)
        
        df['area'] = i + 1
        
        df['Year'] = df['Year'].str.replace('<tt><pre>', '', regex=False)
        
        df = df.drop(df[df['Year'] == '</pre></tt>'].index)

        df.drop('empty', axis = 1, inplace=True)
        
        df['Year'] = df['Year'].astype(int)

        df['Week'] = df['Week'].astype(int)
        
        data_frame = pd.concat([data_frame, df], ignore_index=True)
    
    return data_frame

directory = "lab2"
vhi_data_frame = read_files(directory)

print(vhi_data_frame)
```

- Ряд VHI для області за вказаний рік;

- Пошук екстремумів (min та max) для вказаних областей та
років;

```python
def get_vhi_series_and_extremes(data_frame, area, year):
    subset = data_frame[(data_frame['area'] == area) & (data_frame['Year'] == year)]
    vhi_series = subset[['Week', 'VHI']]
    min_vhi = vhi_series['VHI'].min()
    max_vhi = vhi_series['VHI'].max()
    return vhi_series, min_vhi, max_vhi


area = 2  
year = 1982

vhi_series, min_vhi, max_vhi = get_vhi_series_and_extremes(vhi_data_frame, area, year)

print("Ряд VHI для області за вказаний рік:")
print(vhi_series)
print("\nMin VHI:")
print(min_vhi)
print("Max VHI:")
print(max_vhi)
```

- Ряд VHI за вказаний діапазон років для вказаних областей;

```python
def get_vhi_series_by_year_range(data_frame, areas, start_year, end_year):
    subset = data_frame[(data_frame['area'].isin(areas)) & (data_frame['Year'] >= start_year) & (data_frame['Year'] <= end_year)]
    return subset[['Year', 'VHI', 'area']]

start_year = 2000  
end_year = 2003
areas = [1, 2, 5]  

vhi_series_range = get_vhi_series_by_year_range(vhi_data_frame, areas, start_year, end_year)
print("\nРяд VHI за вказаний діапазон років для вказаних областей:")
print(vhi_series_range)
```

o Для всього набору даних виявити роки, протягом яких
екстремальні посухи торкнулися більше вказаного відсотка
областей по Україні (20% областей - 5 областей з 25).
Повернути роки, назви областей з екстремальними посухами та
значення VHI;

o Аналогічно для помірних посух

```python
def drought_years(data_frame, percentage, vhi_threshold, mode='extreme'):
    years = data_frame['Year'].unique()
    num_areas = len(data_frame['area'].unique())
    threshold = (percentage / 100) * num_areas
    drought_years = []

    for year in years:
        yearly_data = data_frame[data_frame['Year'] == year]
        if mode == 'extreme':
            drought_areas = yearly_data[yearly_data['VHI'] < vhi_threshold]
        elif mode == 'moderate':
            drought_areas = yearly_data[(yearly_data['VHI'] >= 15) & (yearly_data['VHI'] < vhi_threshold)]
        else:
            continue
        
        if len(drought_areas['area'].unique()) > threshold:
            areas = drought_areas['area'].unique()
            vhi_values = {area: drought_areas[drought_areas['area'] == area]['VHI'].tolist() for area in areas}
            drought_years.append({
                'Year': year,
                'Areas': areas.tolist(),
                'VHI': vhi_values
            })
    
    return drought_years

# для екстремальних посух
drought_years_extreme = drought_years(vhi_data_frame, 10, 15, mode='extreme')
print("\nРоки з екстремальними посухами (більше 10% областей з VHI < 15):")
for record in drought_years_extreme:
    print(record)

# для помірних посух
drought_years_moderate = drought_years(vhi_data_frame, 10, 35, mode='moderate')
print("\nРоки з помірними посухами (більше 10% областей з 15 <= VHI < 35):")
for record in drought_years_moderate:
    print(record)
```
